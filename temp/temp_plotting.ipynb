```python
import pandas as pd
import matplotlib.pyplot as plt

results = pd.read_csv("./train_results.csv", index_col=[0])
results
```

```python
import matplotlib.pyplot as plt

# Assuming you have a pandas DataFrame called 'results' with index as model names and columns 'accuracy', 'precision', 'recall', and 'fscore'

# Set the figure size
plt.figure(figsize=(50, 6))

# Set the range for the x-axis
x_min = 0
x_max = 1

# Plot the horizontal bar chart
results.plot.barh(xlim=(x_min, x_max), alpha=0.9)

# Set the y-axis label
plt.ylabel("Models")

# Set the x-axis label
plt.xlabel("Metrics")

# Set the title of the plot
plt.title("Model Performance")

# Show the plot
plt.show()
```
